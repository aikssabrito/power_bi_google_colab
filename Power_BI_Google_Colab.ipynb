## Importando as bibliotecas

```python
from powerbiclient import QuickVisualize, get_dataset_config, Report
from powerbiclient.authentication import DeviceCodeLoginAuthentication

import pandas as pd
```

## Importando e processando os dados

```python
# Importando o arquivo CSV para um Dataframe Pandas
df = pd.read_csv('Financial Sample.csv')

# Mostrando as 5 primeiras linhas do Dataframe
df.head()
```

```python
# Removendo colunas, filtrando os produtos com mais de 1000 unidades vendidas e criando uma coluna informando se o produto teve ou não disconto
df = df.drop(['Month Number', 'Month Name', 'Year'], axis=1)
df = df.loc[df['Units Sold'] > 1000]
df = df.fillna('')
df['Discounted'] = df['Discount Band'] != ''

# Mostrando as 5 primeiras linhas do Dataframe após as alterações
df.head()
```

## Criando a visualização

- Aqui, é necessário fazer a autenticação junto ao Power BI.
- Basta acessar o link fornecido e digitar o código mostrado abaixo.

```python
# Criando o relatório do Power BI a partir do Dataframe
PBI_visualize = QuickVisualize(get_dataset_config(df))
```

```python
# Habilitando o gerenciador de widgets personalizados no Colab
from google.colab import output
output.enable_custom_widget_manager()
```

```python
# Definindo o tamanho do relatório
PBI_visualize.set_size(720, 1280)

# Renderizando o relatório
PBI_visualize
```
